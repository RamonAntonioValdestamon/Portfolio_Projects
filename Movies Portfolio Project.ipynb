```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("movies.csv")
```

```python
df.head()
```

```python
df.count()
```

```python
for col in df.columns:
    num_nullvals = np.sum(df[col].isnull())
    print(col, '-', num_nullvals)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['year'] = df['year'].astype('str')
```

```python
# Separate year and country from month and day
df[['temp_str','year_temp']] = df['released'].str.split(', ', expand = True)
```

```python
# Separate year and country
df[['year released', 'country released']] = df['year_temp'].str.split(' ', n = 1, expand = True)
```

```python
# Separate month and day
df[['month released', 'day released']] = df['temp_str'].str.split(' ', n = 1, expand = True)
```

```python
# Remove unnecessary columns
df = df.drop(['temp_str','year_temp', 'year'], axis = 1)
```

```python
df.head()
```

```python
df.sort_values(['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Correlation between gross revenue and budget

plt.figure(figsize=(8,6))
plt.scatter(df.budget, df.gross, s = 5)
plt.title("Film Budget v.s. Gross Earnings")
plt.xlabel("Film Budget")
plt.ylabel("Gross Earnings")
```

```python
df['budget'].corr(df['gross'])
```

```python
# Plot linear regression
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws={"color":"red"}, line_kws={'color':'blue'})
```

```python
# Correlations between numerical values
df.corr(method = 'pearson', numeric_only = True)
```

```python
# Correlation Matrix between Numerical Movie Factors
sns.heatmap(df.corr(method = 'pearson', numeric_only = True), annot = True, cmap = 'viridis', cbar = False)
plt.title("Correlation Matrix between Numerical Movie Factors")
```

```python
# Adding non-numerical variables to the matrix by dummy variable conversion
df_dummies = df
for col in df_dummies.columns:
    if (df_dummies[col].dtype == 'object'):
        df_dummies[col] = df_dummies[col].astype('category')
        df_dummies[col] = df_dummies[col].cat.codes

df_dummies.head()
```

```python
df_dummies = df_dummies.drop(['month released', 'day released', 'country released'], axis = 1)
```

```python
# Correlation for all variables/factors
plt.figure(figsize = (12,12))
sns.heatmap(df_dummies.corr(method = 'pearson', numeric_only = True), annot = True, cmap = 'viridis', cbar = False)
plt.title("Correlation Matrix for all movie factors")
```

```python
# Unstacking the correlation matrix
corr_pairs = df_dummies.corr().unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr_pairs = sorted_pairs[(sorted_pairs) > 0.5]
high_corr_pairs
```
